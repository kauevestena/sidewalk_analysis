```python
import plotly.express as px
import geopandas as gpd
```

```python
gdf = gpd.read_file('../all_neighborhoods_crossing_analysis.geojson')
```

```python
protoblocks = gpd.read_file('../all_neighborhoods_block_analysis.geojson')
```

```python
# the relative amount 
for neighborhood in list(protoblocks['neighborhood'].unique()) :
    selec = protoblocks.loc[protoblocks['neighborhood'] == neighborhood]
    selec['number_crossings'].value_counts().to_csv(f'number_crossings_{neighborhood}.csv')
```

```python
px.histogram(gdf, 
             x='number_kerbs', 
            #  y="tip", 
             color="neighborhood",
                   marginal="box", # or violin, rug
                #    hover_data=df.columns
                # range_x=(0,30),
               #  width = 900,
               #  nbins=10,
                   )
```
